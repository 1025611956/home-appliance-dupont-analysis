```python
# ==========================
# 7家家电零售企业杜邦分析（最终稳定版）
# 功能：读取Excel → 统计分析 → 输出可打开Excel + 4张高清图
# ==========================
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# -------------------------- 1. 读取你的真实数据 --------------------------
df = pd.read_excel("7家家电零售企业_杜邦分析总表_2020-2024.xlsx")

# 商业模式（自动补齐）
df["商业模式"] = df["企业名称"].map({
    "美的集团":"制造", "格力电器":"制造", "海尔智家":"制造",
    "京东集团":"电商", "阿里巴巴":"电商",
    "苏宁易购":"传统零售", "国美零售":"传统零售"
})

# -------------------------- 2. 统计分析（高质量） --------------------------
# 描述统计
desc = df.groupby("企业名称")[["ROE(%)","销售净利率(%)","总资产周转率","权益乘数"]].agg(["mean","std"]).round(4)

# 相关系数 + 显著性
corr = df[["ROE(%)","销售净利率(%)","总资产周转率","权益乘数"]].corr().round(4)
p_value = df[corr.columns].corr(method=lambda x,y: stats.pearsonr(x,y)[1]).round(4)

# 回归
X = df[["销售净利率(%)","总资产周转率","权益乘数"]]
y = df["ROE(%)"]
model = LinearRegression().fit(X, y)
reg = pd.DataFrame([{
    "截距": round(model.intercept_,4),
    "销售净利率系数": round(model.coef_[0],4),
    "总资产周转率系数": round(model.coef_[1],4),
    "权益乘数系数": round(model.coef_[2],4),
    "R²": round(model.score(X,y),4)
}])

# 方差分析
g1 = df[df["商业模式"]=="制造"]["ROE(%)"]
g2 = df[df["商业模式"]=="电商"]["ROE(%)"]
g3 = df[df["商业模式"]=="传统零售"]["ROE(%)"]
f,p = stats.f_oneway(g1,g2,g3)
anova = pd.DataFrame({
    "F值": [round(f,4)],
    "p值": [round(p,4)],
    "显著性": ["p<0.05显著" if p<0.05 else "不显著"]
})

# 增长率
growth = df.groupby("企业名称")["ROE(%)"].apply(lambda x: x.pct_change().mean()*100).round(2)
growth = growth.reset_index()
growth.columns = ["企业名称","ROE年均增长率(%)"]

# -------------------------- 3. 输出 4 张独立高清图 --------------------------
# 1 箱线图
plt.figure(figsize=(10,5))
df.boxplot(column="ROE(%)", by="商业模式", vert=False)
plt.title("ROE按商业模式分布")
plt.suptitle("")
plt.tight_layout()
plt.savefig("1_ROE箱线图.png", dpi=300, bbox_inches="tight")
plt.close()

# 2 相关性热力图
plt.figure(figsize=(8,6))
im = plt.imshow(corr, cmap="RdBu_r", vmin=-1, vmax=1)
plt.colorbar(im)
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title("核心指标相关性")
plt.tight_layout()
plt.savefig("2_相关性热力图.png", dpi=300)
plt.close()

# 3 ROE趋势
plt.figure(figsize=(12,5))
for name in df["企业名称"].unique():
    sub = df[df["企业名称"]==name]
    plt.plot(sub["年份"], sub["ROE(%)"], marker="o", linewidth=2, label=name)
plt.legend()
plt.title("ROE年度趋势")
plt.tight_layout()
plt.savefig("3_ROE年度趋势图.png", dpi=300)
plt.close()

# 4 回归拟合
plt.figure(figsize=(8,5))
plt.scatter(y, model.predict(X), alpha=0.7)
plt.plot([y.min(),y.max()], [y.min(),y.max()], "r--")
plt.title(f"回归拟合 R²={model.score(X,y):.3f}")
plt.tight_layout()
plt.savefig("4_回归拟合图.png", dpi=300)
plt.close()

# -------------------------- 4. 输出 100% 可打开的标准Excel --------------------------
with pd.ExcelWriter("家电企业杜邦分析结果.xlsx", engine="openpyxl") as f:
    df.to_excel(f, sheet_name="原始数据", index=False)
    desc.to_excel(f, sheet_name="描述统计")
    corr.to_excel(f, sheet_name="相关系数")
    p_value.to_excel(f, sheet_name="显著性P值")
    reg.to_excel(f, sheet_name="回归结果", index=False)
    anova.to_excel(f, sheet_name="方差分析", index=False)
    growth.to_excel(f, sheet_name="ROE增长率", index=False)

# -------------------------- 完成 --------------------------
print("✅ 全部完成！")
print("📊 Excel 可打开：家电企业杜邦分析结果.xlsx")
print("🖼️  4 张高清图已生成")
```

```python

# 功能：读取数据→剔除异常→相关性→回归→输出结果
# ==========================
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')

# 1. 读取原始数据
df = pd.read_excel("7家家电零售企业_杜邦分析总表_2020-2024.xlsx")

# 2. 剔除苏宁易购、国美零售（异常值企业）
df_filtered = df[~df["企业名称"].isin(["苏宁易购", "国美零售"])].reset_index(drop=True)
print("✅ 剔除完成，剩余企业：", df_filtered["企业名称"].unique().tolist())
print("="*60)

# 3. 核心指标定义
indicators = ["ROE(%)", "销售净利率(%)", "总资产周转率", "权益乘数"]

# 4. Pearson相关性分析（输出相关系数+P值，无美化）
corr_matrix = df_filtered[indicators].corr().round(4)
p_matrix = df_filtered[indicators].corr(method=lambda x, y: stats.pearsonr(x, y)[1]).round(4)

print("📊 剔除异常值后 相关性矩阵（真实结果）：")
print(corr_matrix)
print("\n📊 相关性显著性P值矩阵：")
print(p_matrix)
print("="*60)

# 5. 多元线性回归（ROE为因变量，三因子为自变量）
X = df_filtered[["销售净利率(%)", "总资产周转率", "权益乘数"]]
y = df_filtered["ROE(%)"]
model = LinearRegression(fit_intercept=True)  # 强制保留截距，避免误判
model.fit(X, y)

# 6. 整理回归结果（严格对应列名，无错位）
reg_result = pd.DataFrame({
    "截距": [round(model.intercept_, 4)],
    "销售净利率系数": [round(model.coef_[0], 4)],
    "总资产周转率系数": [round(model.coef_[1], 4)],
    "权益乘数系数": [round(model.coef_[2], 4)],
    "R²": [round(model.score(X, y), 4)]
})

print("📈 剔除异常值后 多元线性回归结果（真实结果）：")
print(reg_result.to_string(index=False))  # 隐藏行索引，彻底解决错位
print("="*60)

# 7. 输出Excel（无冗余，列名严格对应）
with pd.ExcelWriter("剔除异常值后严谨分析结果.xlsx", engine="openpyxl") as writer:
    df_filtered.to_excel(writer, sheet_name="01_剔除后原始数据", index=False)
    corr_matrix.to_excel(writer, sheet_name="02_相关系数矩阵")
    p_matrix.to_excel(writer, sheet_name="03_相关性P值矩阵")
    reg_result.to_excel(writer, sheet_name="04_回归分析结果", index=False)

print("✅ 分析完成，结果已保存至：剔除异常值后严谨分析结果.xlsx")
```
